```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import time
import requests
from config import api_key
```

```python
#read
weatherpy_data = pd.read_csv("../WeatherPy/data/weather_cities.csv")
weatherpy_data
```

```python
gmaps.configure(api_key)
locations = weatherpy_data[["Lat", "Lng"]]
humidity = weatherpy_data["Humidity"]

fig = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
maximum_intensity = np.max(humidity)

heat_lay = gmaps.heatmap_layer(locations, weights = humidity, dissipating=False, max_intensity=100, point_radius=3)

fig.add_layer(heat_lay)
fig
```
